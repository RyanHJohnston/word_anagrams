```python
words_bylength = collections.defaultdict(list)
```

```python
for string in wordclean:
    length = len(string)
    if length in words_bylength:
        words_bylength[length].append(string)
    else:
        words_bylength[length] = [string]
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
def anagram(word):
    return [w for w in wordclean if signature(w) == signature(word)]
```

```python
import collections
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(word):
    return words_bysig[signature(word)]
```

```python
anagrams_bylength = collections.defaultdict(dict)
```

```python
# key is the length of each class of words having same length
# value is total number of anagrams in each class of words
anagrams_bylength_total = collections.defaultdict(float)

for length, words in words_bylength.items():
    anagram_counts = collections.defaultdict(float)

    for word in words:
        sorted_word = ''.join(sorted(word))
        anagram_counts[sorted_word] += 1

    total_anagrams = sum(count - 1 for count in anagram_counts.values() if count > 1)
    anagrams_bylength_total[length] = total_anagrams
```
